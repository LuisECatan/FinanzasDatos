```python
import pandas as pd
import numpy as np


np.random.seed(42)

clientes = pd.DataFrame({
'ID_Cliente': range(1, 101),
'Nombre': ['Cliente_' + str(i) for i in range(1, 101)],
'Edad': np.random.randint(18, 70, size=100),
'Género': np.random.choice(['Hombre', 'Mujer'], size=100),
'Tipo_Cuenta': np.random.choice(['Ahorro', 'Corriente', 'Inversión'], size=100),
'Saldo': np.random.uniform(1000, 50000, size=100).round(2)
})

# Datos de transacciones
transacciones = pd.DataFrame({
'ID_Transacción': range(1, 501),
'ID_Cliente': np.random.choice(range(1, 101), size=500),
'Monto': np.random.uniform(100, 10000, size=500).round(2),
'Fecha': pd.date_range(start='2023-01-01', periods=500, freq='D')
})
```
